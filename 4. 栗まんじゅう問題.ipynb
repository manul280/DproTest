【問題】栗まんじゅうが太陽系を覆う日

栗まんじゅうの体積($m^3$)：$0.06 * 0.04 * 0.03$  
太陽系の体積：$太陽系半径^2*\pi$  
太陽系半径($m$)：$1.575*10**(12+3)/2$  ※直径1.575光年（15兆km）

```python
#@@@ 太陽系の体積
# 太陽系半径
half=1.575*10**(12+3)/2
# 太陽系の体積：半径×半径×円周率×4
SUN_WORLD=math.pi*(half**2)
```

```python
def kurifill_time(fill):
    """栗まんじゅうで指定体積を覆う時間計算関数。

    Parameteres
    -----------
    fill: int
        覆う対象の体積

    Returns
    -------
    time_cnt : int
        栗まんじゅうが覆う増殖回数
    kuri_value : int
        栗まんじゅうが増殖する数(リスト)
    """
    KURIMAN=0.06*0.04*0.03
    time_cnt = 1
    kuri_full=KURIMAN
    kuri_value=[]
    while True:
        if fill < kuri_full:
            break
        kuri_full *=2
        kuri_value.append([time_cnt * 5, kuri_full])
        time_cnt +=1
    return time_cnt * 5,kuri_value
```

```python
# 太陽系体積
SUN_WORLD=1.40927*10**18
time,kazu=kurifill_time(SUN_WORLD)
print(f"栗まんじゅうが太陽系を覆う時間: {time}分後")
```

```python
x=[x[0] for x in kazu]
y=[x[1] for x in kazu]
# グラフを表示
plt.title("栗まんじゅうが覆う時間<太陽系>")
plt.xlabel("時間(分)")
plt.ylabel("栗まんじゅう体積(㎡)")
plt.tick_params(labelsize=15) # 軸の値に関する設定を行う
plt.plot(x, y,color='orange') 
plt.show()
```
