```python
import numpy as np
import matplotlib.pyplot as plt
```

【問題1】乱数の作成

```python
np.random.seed(0)
# 平均
mu=[-3,0]
# 分散共分散行列
sigma=[[1.0,0.8],[0.8,1.0]]
# 2次元正規乱数を1万個生成
values0 = np.random.multivariate_normal(mu,  sigma, 500)
```

```python
values0.shape,type(values0)
```

【問題2】散布図による可視化

```python
plt.scatter(values0[:,0],values0[:,1])
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

【問題3】ヒストグラムによる可視化

```python
plt.subplot(211,title=("x1"))
plt.hist(values0[:,0], bins=100)
plt.xlim(-6,4)

plt.subplot(212,title=("x2"))
plt.hist(values0[:,1], bins=100)
plt.xlim(-6,4)
plt.show()
```

【問題4】データの追加

```python
np.random.seed(0)
# 平均
mu=[0,-3]
# 分散共分散行列
sigma=[[1.0,0.8],[0.8,1.0]]
# 2次元正規乱数を1万個生成
values1 = np.random.multivariate_normal(mu,  sigma, 500)
```

```python
values1.shape,type(values1)
```

```python
plt.scatter(values0[:,0],values0[:,1])
plt.scatter(values1[:,0],values0[:,1])
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

【問題5】データの結合

```python
values2 = np.concatenate([values0,values1],axis=0)
values2
```

```python
values2.shape,type(values2)
```

【問題6】ラベル付け

```python
label0 = np.zeros(500)
label1 = np.ones(500)
label2 = np.concatenate([label0,label1],axis=0)
values3 = np.insert(values2,2,label2,axis=1)
values3
```
