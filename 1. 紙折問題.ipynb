【問題1】べき乗の算術演算子を使用した実装

```python
# 紙を1回折った時の厚さを計算するコード
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2**43
print("厚さ： {}メートル".format(folded_thickness))
```

【問題2】単位の変換

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2**43
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
```

【問題3】for文を使用して作成

```python
THICKNESS = 0.00008
folded_thickness=THICKNESS
for i in range(43):
    folded_thickness *=2
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
```

【問題4】計算時間の比較

```python
import time
start = time.time()
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2**43
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

【問題5】リストへの保存

```python
THICKNESS = 0.00008
folded_thickness=THICKNESS
thicknesses=[THICKNESS]
for i in range(43):
    folded_thickness *=2
    thicknesses.append(folded_thickness)
print(thicknesses)
print(len(thicknesses))
```

【問題6】折れ線グラフの表示

```python
import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(thicknesses)
plt.show()
```
