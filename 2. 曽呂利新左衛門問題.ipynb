【問題3】貰える米で何人が何日生活できるか

### 成人男性：2200カロリー/日 = 15000000粒
白米1合（約150g）当たりのカロリー：約513kcal  
150*2200=330000g/日  
千粒重：平均22g ※千粒の重さ  
330000*1000/22=15000000
1日必要米粒数：15000000

```python
def life_day(list_total_grains,man):
    """貰える米で何人が何日生活できるかを計算する関数。

    Parameteres
    -----------
    list_total_grains : int
        米粒の数
    man : int        
        人数(int)
    Returns
    -------
    life_day : int
        貰える米で指定人数で生活可能な日数
    """
    LIFE_GRAINS=15000000
    life_day = list_total_grains // man * LIFE_GRAINS
    return life_day
```

```python
# 15日分米粒をもらい3人で何日生活できるか
day = 15
man = 3
lices,lices_total=compute_sorori_shinzaemon(day)
print(f"{man}人で{life_day(lices_total,man):,}日生活できる")
```
