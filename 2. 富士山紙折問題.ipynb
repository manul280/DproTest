【問題1】何回折ったら富士山を超えるか

```python
THICKNESS = 0.00008
MT_FUJI = 3776
folded_thickness=THICKNESS
folded_cnt=1

while True:
    if MT_FUJI < folded_thickness:
        break
    folded_thickness *=2
    folded_cnt +=1

# 回数
print(f"{folded_cnt:,}回折ったら富士山を超える")
```

【問題2】任意の厚さに対応した関数

```python
def folded_thickness(hight,thickness=0.00008):
    """何回折ったら関数指定のを高さをこえるか関数。

    Parameteres
    -----------
    hight: int
        高さ
    thickness: int
        折る前の厚さ(default : 0.00008)

    Returns
    -------
    min_folded : int
        折込んで指定高さをこえる最小回数
    """
    min_folded=1

    while True:
        if hight < thickness:
            break
        thickness *=2
        min_folded +=1

    # 回数
    return min_folded
```

```python
print(f"{folded_thickness(4.0175*10**16):,}回折ったら「最も近い太陽以外の恒星(プロキシマ・ケンタウリ)」を超える")
```

【問題3】必要な紙の長さ

```python
def need_long (folded=1,thickness=0.00008):
    """指定厚さの紙を指定回折るために必要な紙の長さ計算する関数。

    Parameteres
    -----------
    folded : int
        折込回数
    thickness : float        
        紙の厚さ
    Returns
    -------
    life_day : float
        必要な紙の長さ
    """
    need_long=math.pi*thickness * (2**folded+4) * (2**folded-1)/6
    return need_long
```

```python
# Em エクサメートル：1.E-18	
# Ym ヨタメートル：1.E-24
print(f"折込んで「月」を超える紙の長さ：{need_long(folded=folded_thickness(384400000),thickness=384400000)/10**24:,}Ym")
print(f"折込んで「富士山」を超える紙の長さ：{need_long(folded=folded_thickness(3776),thickness=3776)/10**18:,}Em")
print(f"折込んで「最も近い太陽以外の恒星(プロキシマ・ケンタウリ)」を超える紙の長さ：{need_long(folded=folded_thickness(4.0175*10**16),thickness=4.0175*10**16)/10**24:,}Ym")
```
