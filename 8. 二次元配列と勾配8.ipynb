# 二次元配列と勾配問題

【問題1】1次関数  
$y=\frac{1}{2}x+1$

```python
x=[]
y=[]
for i in range(-500,501):
    xx = i/10
    x.append(xx)
    y.append(xx/2+1)
x = np.array(x)
y = np.array(y)
x,x.shape
```

【問題2】配列の結合

```python
xy=np.stack([x,y],axis=1)
xy,xy.shape
xy[:,0],xy[:,1]
```

【問題3】勾配を求める  
$勾配=\frac{yの変化量}{xの変化量}$

image.png

```python
# X:変化量
d_x = x[1:] - x[:-1]
d_x
```

```python
# Y:変化量
d_y = y[1:] - y[:-1]
```

```python
#勾配
gra = d_y/d_x
gra
```

```python
plt.plot(xy[:-1,0],gra)
plt.title("y=1/2x+1の勾配 グラフ")
plt.xlabel("x")
plt.ylabel("勾配")
plt.show()
```

【問題5】Pythonの関数化

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。

    Parameters
    ----------------
    function : function
      求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
      np.arange()の引数と同じ方法で範囲を指定する。

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
      関数の勾配。差をとるため要素は1つ減る
    """
    x=[]
    y=[]
    for i in range(int(x_range[0]*10),int(x_range[1]*10)):
        xx = i/10
        x.append(xx)
        y.append(function(xx))
    x = np.array(x)
    y = np.array(y)
    array_xy = np.stack([x,y],axis=1)

    # X:変化量
    d_x = x[1:] - x[:-1]
    # Y:変化量
    d_y = y[1:] - y[:-1]
    #勾配
    gradient = d_y/d_x
    return array_xy, gradient
```

### $y=\frac{1}{2}x+1$

```python
def function1(array_x):
    array_y = array_x/2 + 1
    return array_y

array_xy1, gradient1 = compute_gradient(function1)
print(f"y=1/2x+1 勾配：{gradient1}")
```

```python
plt.plot(array_xy1[:,0],array_xy1[:,1])
plt.title("y=1/2x+1 グラフ")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(array_xy1[:-1,0],gradient1)
plt.title("y=1/2x+1の勾配 のグラフ")
plt.xlabel("x")
plt.ylabel("勾配")
plt.show()
```

### $y=2x^2+2^x$

```python
def function3(array_x):
    array_y = 2*array_x**2 + 2**array_x
    return array_y

array_xy3, gradient3 = compute_gradient(function3)
print(f"y=2x^2+2^x 勾配：{ gradient3}")
```

```python
plt.plot(array_xy3[:,0],array_xy3[:,1])
plt.title(" y=2x^2+2^x グラフ")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
